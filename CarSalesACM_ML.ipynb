```python
# ML Code snippet 1
# Setting the noise/redundant/missing data variables
ignored_cols  = ['AccountMainSysID', 'AccountID', 'FullName', 'SaleDate',
                'VendorNumber',  'BuyerName', 'BuyerNumber','CMACount','VehicleMainSysID', 'LoanOriginationDate',
                 'StockNumber','VinNumber','PurchaseIntent', 'VehicleExpenses', 'DatePurchased', 'EstimatedPayoffDate',
                 'VehicleExpense', 'CustomerRepair', 'ServiceContractRepair', 'RepairToRetail', 'TotalExpenses','BuyerName', 'BuyerNumber',
                 'MaxTerm', 'ActualTerm', 'LotBuyInDate', 'CompanyBuyInDate', 'SaleType', 'SuggestedPurchasePrice', 'GrossTradeInAllowance',
                 'TradeInPayoff',]






```

```python
# ML Code snippet 2
# Setting categorical cols
cat_cols = [ 'LotSysID', 'LotID', 'Area', 'Region',  'Trade',
       'ModelYear', 'Make', 'Model', 'Color', 'Cylinder', 'Transmission', 'DriveTrain',
       'Trim', 'Body', 'VehicleClass', 'VehicleCategory',  'PurchaseType',
       'StateCodeLot']
```

```python
# ML Code snippet 3
# Setting the numerical cols
num_cols = ['SalesPrice','Mileage',
      'DaysOnLot',  'DepositAmount', 'CashDown'
      , 'PurchasedPrice', 'SuggestedSalesPrice']
```

```python
# ML Code Snippet 4
# Setting the dependent/target/label variable
y = 'DaysInCompany'
```

```python
# ML Code snippet 5
# Installing Pycaret for Automated ML
!pip install -U --pre pycaret
```

```python
data.dropna(subset=['DaysInCompany'], inplace=True)
```

```python
# ML Code snippet 6
# Importing Pycaret and setting/configuring the ML experiment
from pycaret.regression import *

regression_model_setup = setup(data,
      target = y,
      categorical_features = cat_cols,
      numeric_features = num_cols,
      ignore_features = ignored_cols)
```
